```python
import os
import numpy as np
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.nn.functional as F
from torch import optim
import torch
from torchvision import datasets
from torchvision import transforms
from torch.utils.data import DataLoader
from tqdm import tqdm
from torch.utils.tensorboard import SummaryWriter
import random  # Add this to fix the error

# CNN Model
class CNN2(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1)
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.dense1 = nn.Linear(32*14*14, 128)
        self.dense2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.conv1(x)
        x = F.relu(x)
        x = self.pool1(x)
        x = x.view(x.shape[0], -1)
        x = self.dense1(x)
        x = self.dense2(x)
        out = F.log_softmax(x, dim=1)
        return out

# Training Class
class Trainer:

    def __init__(self, model, device, criterion, optimizer, training_DataLoader, validation_DataLoader, epochs=100, path2write=None):
        self.model = model
        self.criterion = criterion
        self.optimizer = optimizer
        self.training_DataLoader = training_DataLoader
        self.validation_DataLoader = validation_DataLoader
        self.device = device
        self.epochs = epochs
        self.path2write = path2write
        self.writer_train = SummaryWriter(os.path.join(path2write, "train"))
        self.writer_val = SummaryWriter(os.path.join(path2write, "val"))
        self.check_point_path = os.path.join(path2write, 'check_points')
        os.makedirs(self.check_point_path, exist_ok=True)
        self.training_loss = []
        self.validation_loss = []
        self.training_accuracy = []
        self.validation_accuracy = []

    def run_trainer(self):
        self.model.to(self.device)
        for epoch in range(self.epochs):
            print(f'Epoch {epoch+1}/{self.epochs}')
            # Training
            train_loss, train_accuracy = self._train()
            self.writer_train.add_scalar("Train Loss", train_loss, epoch)
            self.writer_train.add_scalar("Train Accuracy", train_accuracy, epoch)
            # Validation
            val_loss, val_accuracy = self._validate()
            self.writer_val.add_scalar("Val Loss", val_loss, epoch)
            self.writer_val.add_scalar("Val Accuracy", val_accuracy, epoch)
            # Logging
            print(f'Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Train Acc: {train_accuracy:.4f}, Val Acc: {val_accuracy:.4f}')
            if epoch % 10 == 0 or epoch == self.epochs-1:
                torch.save(self.model.state_dict(), os.path.join(self.check_point_path, f'epoch-{epoch}-loss-{val_loss:.4f}.pth'))
        return self.training_loss, self.validation_loss, self.model, self.training_accuracy, self.validation_accuracy

    def _train(self):
        self.model.train()
        train_losses = []
        correct = 0
        total = 0
        for x, y in tqdm(self.training_DataLoader, desc='Training'):
            x, y = x.to(self.device), y.to(self.device)
            self.optimizer.zero_grad()
            output = self.model(x)
            loss = self.criterion(output, y)
            loss.backward()
            self.optimizer.step()
            train_losses.append(loss.item())
            pred = output.argmax(dim=1, keepdim=True)
            correct += pred.eq(y.view_as(pred)).sum().item()
            total += y.size(0)
        accuracy = correct / total
        self.training_loss.append(np.mean(train_losses))
        self.training_accuracy.append(accuracy)
        return np.mean(train_losses), accuracy

    def _validate(self):
        self.model.eval()
        valid_losses = []
        correct = 0
        total = 0
        with torch.no_grad():
            for x, y in tqdm(self.validation_DataLoader, desc='Validation'):
                x, y = x.to(self.device), y.to(self.device)
                output = self.model(x)
                loss = self.criterion(output, y)
                valid_losses.append(loss.item())
                pred = output.argmax(dim=1, keepdim=True)
                correct += pred.eq(y.view_as(pred)).sum().item()
                total += y.size(0)
        accuracy = correct / total
        self.validation_loss.append(np.mean(valid_losses))
        self.validation_accuracy.append(accuracy)
        return np.mean(valid_losses), accuracy

# Hyperparameters and data loading
batch_size = 1020
transform = transforms.Compose([transforms.ToTensor()])

# Ensure paths to datasets are correct and the dataset folder exists
dataset1 = datasets.MNIST('./data', train=True, download=True, transform=transform)
random.shuffle(dataset1.targets.numpy())  # Randomizing labels

dataset2 = datasets.MNIST('./data', train=False, transform=transform)

training_DataLoader = DataLoader(dataset1, batch_size=batch_size, shuffle=True)
validation_DataLoader = DataLoader(dataset2, batch_size=batch_size, shuffle=False)

# Choose the device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Loss function, optimizer, and training configurations
loss_fn = nn.CrossEntropyLoss()
lr = 1e-4
epochs = 100

# Ensure the output directory exists
path2write = "./CNN_training_output"
os.makedirs(path2write, exist_ok=True)

# Model, optimizer, and trainer setup
model = CNN2()
optimizer = torch.optim.Adam(model.parameters(), lr=lr)

# Initialize and run the trainer
trainer = Trainer(model=model, device=device, criterion=loss_fn, optimizer=optimizer, 
                  training_DataLoader=training_DataLoader, validation_DataLoader=validation_DataLoader, 
                  epochs=epochs, path2write=path2write)
training_loss, validation_loss, model, training_accuracy, validation_accuracy = trainer.run_trainer()

# Plot results
fig = plt.figure(figsize=(10,5))
plt.plot(training_loss, label='Training Loss')
plt.plot(validation_loss, label='Validation Loss')
plt.legend()
plt.title('Training and Validation Loss')
plt.show()
fig.savefig(os.path.join(path2write, 'Random_Labels_Loss.png'))
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.nn.functional as F
import torch
from torch import optim
from torchvision import datasets
from torchvision import transforms
from torch.utils.data import DataLoader
from tqdm import tqdm
from sklearn.metrics import confusion_matrix, precision_score, recall_score

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

class Trainer:
    def __init__(self, model, device, criterion, optimizer, training_DataLoader, validation_DataLoader, epochs=100, path2write=None):
        self.model = model
        self.criterion = criterion
        self.optimizer = optimizer
        self.training_DataLoader = training_DataLoader
        self.validation_DataLoader = validation_DataLoader
        self.device = device
        self.epochs = epochs
        self.path2write = path2write
        self.training_loss = []
        self.validation_loss = []
        self.training_accuracy = []
        self.validation_accuracy = []

    def run_trainer(self):
        self.model.to(self.device)
        for epoch in range(self.epochs):
            print(f'Epoch {epoch+1}/{self.epochs}')
            train_loss, train_accuracy = self._train()
            val_loss, val_accuracy = self._validate()
            print(f'Epoch {epoch+1} - Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Train Acc: {train_accuracy:.4f}, Val Acc: {val_accuracy:.4f}')
        sensitivity = self.sensitivity()
        return train_loss, train_accuracy, val_loss, val_accuracy, sensitivity

    def _train(self):
        self.model.train()
        train_losses = []
        correct = 0
        total = 0
        for x, y in tqdm(self.training_DataLoader, desc='Training'):
            x, y = x.to(self.device), y.to(self.device)
            self.optimizer.zero_grad()
            output = self.model(x)
            loss = self.criterion(output, y)
            loss.backward()
            self.optimizer.step()
            train_losses.append(loss.item())
            pred = output.argmax(dim=1, keepdim=True)
            correct += pred.eq(y.view_as(pred)).sum().item()
            total += y.size(0)
        accuracy = correct / total
        return np.mean(train_losses), accuracy

    def _validate(self):
        self.model.eval()
        valid_losses = []
        correct = 0
        total = 0
        with torch.no_grad():
            for x, y in tqdm(self.validation_DataLoader, desc='Validation'):
                x, y = x.to(self.device), y.to(self.device)
                output = self.model(x)
                loss = self.criterion(output, y)
                valid_losses.append(loss.item())
                pred = output.argmax(dim=1, keepdim=True)
                correct += pred.eq(y.view_as(pred)).sum().item()
                total += y.size(0)
        accuracy = correct / total
        return np.mean(valid_losses), accuracy

    def sensitivity(self):
        num = 0
        FNorm = 0
        for p in self.model.parameters():
            if p.grad is not None:
                num += 1
                FNorm += torch.linalg.norm(p.grad).cpu().numpy()
        return FNorm / num if num > 0 else 0

class CNN2(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1)
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.dense1 = nn.Linear(32*14*14, 128)
        self.dense2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.conv1(x)
        x = F.relu(x)
        x = self.pool1(x)
        x = x.view(x.shape[0], -1)
        x = self.dense1(x)
        x = self.dense2(x)
        return F.log_softmax(x, dim=1)

# Hyperparameters and configurations
batch_sizes = [8, 32, 128, 512, 2048, 4096]
train_loss_ = []
val_loss_ = []
train_acc_ = []
val_acc_ = []
sensitivity_ = []
lr = 1e-4
epochs = 15
path2write = "./sensitivity_report"

# Ensure the directory exists
os.makedirs(path2write, exist_ok=True)

# Loop through batch sizes
for batch_size in tqdm(batch_sizes):
    transform = transforms.Compose([transforms.ToTensor()])
    dataset1 = datasets.MNIST('./data', train=True, download=True, transform=transform)
    dataset2 = datasets.MNIST('./data', train=False, transform=transform)

    training_DataLoader = DataLoader(dataset1, batch_size=batch_size, shuffle=True)
    validation_DataLoader = DataLoader(dataset2, batch_size=batch_size, shuffle=True)

    model = CNN2().to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn = nn.CrossEntropyLoss()

    trainer = Trainer(model=model, device=device, criterion=loss_fn, optimizer=optimizer,
                      training_DataLoader=training_DataLoader, validation_DataLoader=validation_DataLoader,
                      epochs=epochs, path2write=path2write)
    
    training_loss, training_accuracy, validation_loss, validation_accuracy, sensitivity = trainer.run_trainer()
    
    train_loss_.append(training_loss)
    val_loss_.append(validation_loss)
    train_acc_.append(training_accuracy)
    val_acc_.append(validation_accuracy)
    sensitivity_.append(sensitivity)

# Plot Batch Size vs Loss and Sensitivity
fig, ax1 = plt.subplots(figsize=(15, 5))
ax1.plot(batch_sizes, train_loss_, 'b-', label='Train Loss')
ax1.plot(batch_sizes, val_loss_, 'b--', label='Validation Loss')
ax1.set_xlabel('Batch Size')
ax1.set_ylabel('Loss', color='b')
ax1.set_title('Batch Size vs Loss vs Sensitivity')
ax1.legend()

ax2 = ax1.twinx()
ax2.plot(batch_sizes, sensitivity_, 'r-', label='Sensitivity')
ax2.set_ylabel('Sensitivity', color='r')
ax2.legend(loc='upper right')

fig.tight_layout()
fig.savefig(os.path.join(path2write, 'sensitivity_loss.png'))

# Plot Batch Size vs Accuracy and Sensitivity
fig, ax1 = plt.subplots(figsize=(15, 5))
ax1.plot(batch_sizes, train_acc_, 'b-', label='Train Accuracy')
ax1.plot(batch_sizes, val_acc_, 'b--', label='Validation Accuracy')
ax1.set_xlabel('Batch Size')
ax1.set_ylabel('Accuracy', color='b')
ax1.set_title('Batch Size vs Accuracy vs Sensitivity')
ax1.legend()

ax2 = ax1.twinx()
ax2.plot(batch_sizes, sensitivity_, 'r-', label='Sensitivity')
ax2.set_ylabel('Sensitivity', color='r')
ax2.legend(loc='upper right')

fig.tight_layout()
fig.savefig(os.path.join(path2write, 'sensitivity_accuracy.png'))
plt.show()
```
